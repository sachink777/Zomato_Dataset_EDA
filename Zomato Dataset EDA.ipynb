```python
plt.figure(figsize=(50,40))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False,cmap='viridis')
```

```python
df_country = pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

observation from above analysis - 
more than 94% trasaction of zomato are from india after that USA and UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
ratings
```

Observation - 
1. When rating is between 4.5 to 4.9  ---------> Excellent
2. When rating is between 4.5 to 4.0  ---------> Very good
3. When rating is between 3.9 to 3.5  ---------> good
4. When rating is between 3.4 to 2.5  ---------> Average
5. When rating is between 2.4 to 1.8  ---------> poor

```python
ratings.head()
```

```python
plt.figure(figsize=(50,40))
sns.barplot(x='Aggregate rating',y='rating count', data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

Observation from above barplot- 
1. not rated count is very high.
2. maximumnumber of rating are between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
